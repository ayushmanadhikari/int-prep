```python
"""Calculate student grades by combining data from many sources.

Using pandas, this script combines data from the:

* Roster
* Homework & Exam grades
* Quiz grades

to calculate final grades for a class.
"""
```

```python
import pandas as pd
import random
```

```python
## Rooster .csv
cols_req = ['ID', 'Name', 'NetID', 'Email Address']
rooster = pd.read_csv('rooster.csv', skipinitialspace=True ,usecols=cols_req)
rooster = rooster.replace(r'\n',' ', regex=True) 
```

```python
rooster.head(10)
```

```python
## Homework .csm
cols_req2 = ['SID', 'First Name', 'Last Name', 'Homework 1', 'Homework 1 - Max Points',
            'Homework 2', 'Homework 2 - Max Points']
homework = pd.read_csv('homework.csv', skipinitialspace=True, usecols=cols_req2)
homework = homework.replace(r'\n', ' ', regex=True)
homework.head(10)
```

```python
homework['Homework 1'].fillna(value=0, inplace=True)
homework['Homework 2'].fillna(value=0, inplace=True)
homework['SID'] = homework['SID'].str.upper()
homework['SID'] = homework['SID'].str.strip()
```

```python
rooster['NetID'] = rooster['NetID'].str.strip()
rooster['NetID']
```

```python
final_col = ['Identifier', 'Name', 'Homework', 'Exams', 'Final Score']
final_df = pd.DataFrame(columns=final_col)
```

```python
homework['Full Name'] = homework['Last Name'] + ', ' + homework['First Name']
```

```python
final_df['Identifier'] = rooster['NetID']
```

```python
## Do not run more than once
final_df = final_df.merge(homework[['SID', 'Full Name']], how='inner', right_on='SID', left_on='Identifier')
```

```python
final_df
```

```python
final_df['Name'] = final_df['Full Name']
final_df.drop(['SID', 'Full Name'], axis=1, inplace=True)
```

```python
final_df.head(6)
```

```python
final_df['Homework'] = (homework['Homework 1'] + homework['Homework 2'])/(homework['Homework 1 - Max Points']+ homework['Homework 2 - Max Points'])
```

```python
final_df
```

```python
list1 = []
for i in range(0, len(final_df)):
    list1.append(random.random())
```

```python
exams = pd.Series(list1)
final_df['Exams'] = exams
```

```python
final_df
```

```python
### Now calculating grade 
## we take weighted sum i.e. 0.35 for homework and 0.65 for exams
```

```python
final_df['Final Score'] = final_df['Homework']*0.35 + final_df['Exams']*0.65
```

```python
final_df
```

```python
### For clearing final df
leng=len(final_df.index)
rows = range(0, leng)
final_df.drop(rows, inplace=True)
#final_df.drop(final_df.columns[5], axis=1, inplace=True)
```
